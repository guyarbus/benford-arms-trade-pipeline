## ABSTRACT 

The data cleaning phase aimed to ensure the **reliability**, **consistency**, and **operational relevance** of the dataset prior to analysis and modeling.  

Raw data were systematically inspected to identify missing values, outliers, duplicates, and incoherent records, particularly those resulting from sensor noise, transmission errors, or adversarial perturbations. Domain-specific rules and statistical thresholds were applied to validate measurements and enforce physical plausibility constraints. Missing or corrupted data were handled using controlled imputation strategies or exclusion when necessary to avoid bias.  

This process was essential to reduce noise, enhance signal integrity, and guarantee that downstream analytical and predictive models operate on data aligned with defense and security requirements.

## BEST PRACTICES & PRINCIPLES

- Always keep raw data unchanged (read-only)
- Document every decision and assumption
- Version control your cleaning scripts
- Make it reproducible (seeds, scripts, environments)
- Validate at each step before proceeding
- Visualize before and after each major change
- Domain expertise consultation when uncertain
- Automate repetitive tasks
- Test on sample before applying to full dataset
- Consider ethical implications (bias, fairness, privacy)

## COMMON PITFALLS TO AVOID

- Deleting data without understanding why
- Imputing without investigating missingness patterns
- Removing outliers that are legitimate
- Data leakage (using test set info in cleaning)
- Over-cleaning (removing valuable variance)
- Ignoring domain knowledge
- Not documenting changes
- Cleaning test data differently than training data
- Assuming correlation implies causation in feature selection
- Not validating cleaned data

## 01. INITIAL DATA ASSESSMENT AND DOCUMENTATION

### 01.1 Data provenance

The datasets used in this study originates from the **Stockholm International Peace Research Institute (SIPRI)**, a globally recognized independent research institute dedicated to conflict, armaments, arms control, and disarmament. SIPRI is widely regarded as the gold standard source for arms transfer data in both academic and policy circles.

Specifically, the data was sourced from the **SIPRI Arms Transfers Database**, which tracks the flow of major conventional weapons between countries. It was accessed on [*date of access*] at [https://www.sipri.org/databases/armstransfers](https://www.sipri.org/databases/armstransfers). The dataset covers the period [*e.g. 1950–2023*] and includes variables such as supplier country, recipient country, weapon category, number of units, and the SIPRI Trend Indicator Value (TIV) — a unit specific to SIPRI that measures the volume of transfers based on estimated production costs rather than financial value.

**Data collection methodology:** SIPRI compiles its data from open sources including national defense reports, parliamentary records, customs data, specialist journals, and news sources. The data is therefore secondary and institutional in nature.

**Key limitations to acknowledge:**
- The **TIV is not a monetary value** and should not be interpreted as such — it reflects military capability transfer volume.
- Data may be **incomplete for non-transparent regimes** where arms trade information is classified or deliberately obscured.
- Small arms and light weapons are **not covered** by this database, which focuses on major conventional weapons.

### 01.2 Initial exploration

Upon loading the dataset, a preliminary exploration was conducted to assess its structure, quality, and overall characteristics before any analytical work.

#### Structural overview

The dataset contains [*n*] observations and [*p*] variables. Each row represents a [*bilateral arms transfer / transaction / country-year*]. Key variables include supplier country, recipient country, weapon description, TIV value, order date, and delivery date. Variable types were verified — categorical variables such as country names and weapon categories were distinguished from numerical variables such as TIV values and quantities.

#### Descriptive statistics

Summary statistics were computed for all numerical variables, including measures of central tendency (mean, median) and dispersion (standard deviation, range). This revealed, for instance, that TIV values are highly right-skewed, with a small number of large transfers dominating the distribution — a common pattern in arms trade data.

#### Missing values

A systematic check for missing or null values was performed. Missing data was found primarily in [*e.g. delivery dates, number of units*], which is expected given the opacity of certain transfers. The extent and pattern of missingness were documented to inform later decisions on imputation or exclusion.

#### Temporal and geographic coverage

The dataset spans [*time period*] and covers [*n*] unique supplier countries and [*n*] unique recipient countries, confirming broad global coverage. Distributions over time were visualized to identify any gaps or anomalies in reporting.

#### Anomalies and outliers
